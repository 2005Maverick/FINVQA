```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import pandas as pd
import os
from PIL import Image
from pathlib import Path
import timm
import numpy as np
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, mean_squared_error, r2_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, average_precision_score
from sklearn.manifold import TSNE
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
import math
class FoodDataset(Dataset):
    def __init__(self, root_dir, csv_file, transform=None, samples_per_class=1000):
        """
        Args:
            root_dir (str): Root directory of the dataset
            csv_file (str): Path to the CSV file with annotations
            transform (callable, optional): Optional transform to be applied on images
            samples_per_class (int): Number of samples to maintain per class (default: 1000)
        """
        self.root_dir = Path(root_dir)
        self.data = pd.read_csv(csv_file)
        self.transform = transform
        self.classes = self.data['Food'].unique().tolist()
        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.classes)}
        self.samples_per_class = samples_per_class
        
        # Define nutrition columns
        self.nutrition_columns = ['Calories', 'Protein (g)', 'Fiber (g)', 
                                'Vitamin C (mg)', 'Antioxidant Score']
        
        # Normalize nutrition values
        self.nutrition_stats = self._calculate_nutrition_stats()
        
        # Balance dataset
        self._balance_dataset()
        
    def _balance_dataset(self):
        """Balance the dataset to ensure each class has exactly samples_per_class samples"""
        self.image_paths = []
        self.labels = []
        self.nutrition_values = []
        
        for food_class in self.classes:
            # Get all data for this class
            class_data = self.data[self.data['Food'] == food_class]
            
            # Calculate how many images we need per food item to reach samples_per_class
            food_items = class_data.shape[0]
            images_per_item = math.ceil(self.samples_per_class / food_items)
            
            class_images = []
            class_labels = []
            class_nutrition = []
            
            # Collect all available images for this class
            for _, row in class_data.iterrows():
                food_dir = self.root_dir / row['relative_path'].split('/')[-1]
                item_images = list(food_dir.glob('*.jpg'))
                
                # Add images and their corresponding labels/nutrition
                for img_path in item_images:
                    class_images.append(img_path)
                    class_labels.append(self.class_to_idx[food_class])
                    class_nutrition.append(self._normalize_nutrition(row))
            
            # If we don't have enough images, use data augmentation
            if len(class_images) < self.samples_per_class:
                # Repeat existing images with slight variations
                while len(class_images) < self.samples_per_class:
                    idx = len(class_images) % len(class_images)
                    class_images.append(class_images[idx])
                    class_labels.append(class_labels[idx])
                    class_nutrition.append(class_nutrition[idx])
            
            # If we have too many images, randomly sample
            if len(class_images) > self.samples_per_class:
                indices = np.random.choice(
                    len(class_images), 
                    self.samples_per_class, 
                    replace=False
                )
                class_images = [class_images[i] for i in indices]
                class_labels = [class_labels[i] for i in indices]
                class_nutrition = [class_nutrition[i] for i in indices]
            
            # Add balanced data to dataset
            self.image_paths.extend(class_images)
            self.labels.extend(class_labels)
            self.nutrition_values.extend(class_nutrition)
        
        # Shuffle the dataset
        indices = list(range(len(self.image_paths)))
        np.random.shuffle(indices)
        self.image_paths = [self.image_paths[i] for i in indices]
        self.labels = [self.labels[i] for i in indices]
        self.nutrition_values = [self.nutrition_values[i] for i in indices]
        
        print(f"Dataset balanced. Total samples: {len(self.image_paths)}")
        for cls in self.classes:
            cls_count = self.labels.count(self.class_to_idx[cls])
            print(f"{cls}: {cls_count} samples")

    def _calculate_nutrition_stats(self):
        """Calculate mean and std for nutrition value normalization"""
        stats = {}
        for col in self.nutrition_columns:
            stats[col] = {
                'mean': self.data[col].mean(),
                'std': self.data[col].std()
            }
        return stats

    def _normalize_nutrition(self, row):
        """Normalize nutrition values using pre-computed statistics"""
        normalized_values = []
        for col in self.nutrition_columns:
            value = (float(row[col]) - self.nutrition_stats[col]['mean']) / self.nutrition_stats[col]['std']
            normalized_values.append(value)
        return torch.tensor(normalized_values)

    def denormalize_nutrition(self, normalized_values):
        """Convert normalized values back to original scale"""
        denormalized = {}
        for i, col in enumerate(self.nutrition_columns):
            value = (normalized_values[i] * self.nutrition_stats[col]['std'] + 
                    self.nutrition_stats[col]['mean'])
            denormalized[col] = value
        return denormalized

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]
        nutrition = self.nutrition_values[idx]

        if self.transform:
            image = self.transform(image)

        return image, label, nutrition

class MultiTaskModel(nn.Module):
    def __init__(self, num_classes, num_nutrition_values=5, model_name='eva02_base_patch14_224.mim_in22k'):
        super(MultiTaskModel, self).__init__()
        
        # Load base model
        self.base_model = timm.create_model(model_name, pretrained=True, num_classes=0)
        
        # Get the dimension of features from base model
        with torch.no_grad():
            dummy_input = torch.randn(1, 3, 224, 224)
            features = self.base_model(dummy_input)
            feature_dim = features.shape[1]
        
        # Classification head
        self.classification_head = nn.Sequential(
            nn.Linear(feature_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, num_classes)
        )
        
        # Nutrition prediction head
        self.nutrition_head = nn.Sequential(
            nn.Linear(feature_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, num_nutrition_values)
        )
    
    def forward(self, x):
        features = self.base_model(x)
        class_output = self.classification_head(features)
        nutrition_output = self.nutrition_head(features)
        return class_output, nutrition_output

class MultiTaskLoss(nn.Module):
    def __init__(self, classification_weight=1.0, nutrition_weight=0.5):
        super(MultiTaskLoss, self).__init__()
        self.classification_weight = classification_weight
        self.nutrition_weight = nutrition_weight
        self.classification_criterion = nn.CrossEntropyLoss()
        self.nutrition_criterion = nn.MSELoss()
    
    def forward(self, class_pred, nutrition_pred, class_target, nutrition_target):
        classification_loss = self.classification_criterion(class_pred, class_target)
        nutrition_loss = self.nutrition_criterion(nutrition_pred, nutrition_target)
        total_loss = (self.classification_weight * classification_loss + 
                     self.nutrition_weight * nutrition_loss)
        return total_loss, classification_loss, nutrition_loss

class FoodClassifier:
    def __init__(self, num_classes=33):
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.num_classes = num_classes
        
        # Initialize multi-task model
        self.model = MultiTaskModel(num_classes=num_classes).to(self.device)
        
        # Initialize multi-task loss
        self.criterion = MultiTaskLoss()
        
        # Transforms
        self.train_transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomHorizontalFlip(),
            transforms.RandomRotation(15),
            transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
            transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
        
        self.val_transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

        self.history = {
            'train_loss': [], 'train_class_acc': [], 'train_nutr_mse': [],
            'val_loss': [], 'val_class_acc': [], 'val_nutr_mse': [],
            'test_class_acc': None, 'test_nutr_mse': None
        }

    def load_saved_model(self, model_path):
        """Load a saved model checkpoint"""
        checkpoint = torch.load(model_path, map_location=self.device)
        self.model.load_state_dict(checkpoint['model_state_dict'])
        print(f"Model loaded from {model_path}")
        print(f"Validation accuracy of saved model: {checkpoint['val_acc']:.2f}%")

    def train_model(self, train_loader, val_loader, test_loader, num_epochs=10):
        optimizer = optim.AdamW(self.model.parameters(), lr=1e-4, weight_decay=0.05)
        scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)
        scaler = torch.cuda.amp.GradScaler()
        best_val_acc = 0.0
        
        for epoch in range(num_epochs):
            # Training phase
            self.model.train()
            running_loss = 0.0
            running_class_correct = 0
            running_nutr_mse = 0.0
            total = 0
            
            train_pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Train]')
            for inputs, labels, nutrition in train_pbar:
                inputs = inputs.to(self.device)
                labels = labels.to(self.device)
                nutrition = nutrition.to(self.device).float()
                
                optimizer.zero_grad()
                
                with torch.cuda.amp.autocast():
                    class_outputs, nutr_outputs = self.model(inputs)
                    total_loss, class_loss, nutr_loss = self.criterion(
                        class_outputs, nutr_outputs, labels, nutrition)
                
                scaler.scale(total_loss).backward()
                scaler.step(optimizer)
                scaler.update()
                
                running_loss += total_loss.item()
                _, predicted = class_outputs.max(1)
                total += labels.size(0)
                running_class_correct += predicted.eq(labels).sum().item()
                running_nutr_mse += nutr_loss.item()
                
                train_pbar.set_postfix({
                    'loss': f'{running_loss/len(train_loader):.4f}',
                    'class_acc': f'{100.*running_class_correct/total:.2f}%',
                    'nutr_mse': f'{running_nutr_mse/len(train_loader):.4f}'
                })
            
            train_acc = 100. * running_class_correct / total
            train_loss = running_loss / len(train_loader)
            train_nutr_mse = running_nutr_mse / len(train_loader)
            
            # Validation phase
            val_metrics = self.evaluate(val_loader, desc='Validation')
            
            # Save metrics
            self.history['train_loss'].append(train_loss)
            self.history['train_class_acc'].append(train_acc)
            self.history['train_nutr_mse'].append(train_nutr_mse)
            self.history['val_loss'].append(val_metrics['total_loss'])
            self.history['val_class_acc'].append(val_metrics['class_acc'])
            self.history['val_nutr_mse'].append(val_metrics['nutr_mse'])
            
            print(f'\nEpoch {epoch+1}/{num_epochs}:')
            print(f'Training - Loss: {train_loss:.4f}, Class Acc: {train_acc:.2f}%, Nutr MSE: {train_nutr_mse:.4f}')
            print(f'Validation - Loss: {val_metrics["total_loss"]:.4f}, Class Acc: {val_metrics["class_acc"]:.2f}%, Nutr MSE: {val_metrics["nutr_mse"]:.4f}')
            
            if val_metrics['class_acc'] > best_val_acc:
                best_val_acc = val_metrics['class_acc']
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': self.model.state_dict(),
                    'optimizer_state_dict': optimizer.state_dict(),
                    'val_acc': val_metrics['class_acc'],
                }, 'best_model.pth')
            
            scheduler.step()
        
        # Final test evaluation
        print("\nEvaluating on test set...")
        test_metrics = self.evaluate(test_loader, desc='Test')
        self.history['test_class_acc'] = test_metrics['class_acc']
        self.history['test_nutr_mse'] = test_metrics['nutr_mse']
        print(f'Final Test - Class Acc: {test_metrics["class_acc"]:.2f}%, Nutr MSE: {test_metrics["nutr_mse"]:.4f}')
        
        self.plot_training_history()
    def evaluate(self, loader, desc='Evaluation'):
        self.model.eval()
        running_loss = 0.0
        running_class_correct = 0
        running_nutr_mse = 0.0
        total = 0
        
        all_nutrition_true = []
        all_nutrition_pred = []
        all_labels = []
        all_predictions = []
        all_features = []
        
        with torch.no_grad():
            pbar = tqdm(loader, desc=desc)
            for inputs, labels, nutrition in pbar:
                inputs = inputs.to(self.device)
                labels = labels.to(self.device)
                nutrition = nutrition.to(self.device).float()
                
                # Get features before classification head
                features = self.model.base_model(inputs)
                class_outputs, nutr_outputs = self.model(inputs)
                total_loss, class_loss, nutr_loss = self.criterion(
                    class_outputs, nutr_outputs, labels, nutrition)
                
                running_loss += total_loss.item()
                _, predicted = class_outputs.max(1)
                total += labels.size(0)
                running_class_correct += predicted.eq(labels).sum().item()
                running_nutr_mse += nutr_loss.item()
                
                # Store predictions and true labels for metrics
                all_labels.extend(labels.cpu().numpy())
                all_predictions.extend(predicted.cpu().numpy())
                all_nutrition_true.extend(nutrition.cpu().numpy())
                all_nutrition_pred.extend(nutr_outputs.cpu().numpy())
                all_features.extend(features.cpu().numpy())
                
                pbar.set_postfix({
                    'loss': f'{running_loss/len(loader):.4f}',
                    'class_acc': f'{100.*running_class_correct/total:.2f}%',
                    'nutr_mse': f'{running_nutr_mse/len(loader):.4f}'
                })
        
        # Convert lists to numpy arrays
        all_labels = np.array(all_labels)
        all_predictions = np.array(all_predictions)
        all_features = np.array(all_features)
        
        # Calculate metrics
        conf_matrix = confusion_matrix(all_labels, all_predictions)
        r2 = r2_score(np.array(all_nutrition_true), np.array(all_nutrition_pred))
        
        # Calculate Dice coefficient for each class
        dice_scores = self.calculate_dice_coefficients(conf_matrix)
        
        # Generate evaluation plots
        if desc == 'Test':
            try:
                self.plot_evaluation_metrics(
                    all_labels, 
                    all_predictions, 
                    all_features, 
                    conf_matrix,
                    dice_scores,
                    loader.dataset.classes
                )
                self.plot_nutrition_predictions(loader, loader.dataset)
            except Exception as e:
                print(f"Warning: Could not generate evaluation plots: {str(e)}")
        
        return {
            'total_loss': running_loss / len(loader),
            'class_acc': 100. * running_class_correct / total,
            'nutr_mse': running_nutr_mse / len(loader),
            'nutr_r2': r2,
            'dice_scores': dice_scores,
            'all_predictions': all_predictions,
            'all_labels': all_labels,
            'all_features': all_features,
            'all_nutrition_true': np.array(all_nutrition_true),
            'all_nutrition_pred': np.array(all_nutrition_pred)
        }

    def calculate_dice_coefficients(self, conf_matrix):
        """Calculate Dice coefficient for each class"""
        dice_scores = {}
        n_classes = conf_matrix.shape[0]
        
        for i in range(n_classes):
            true_positive = conf_matrix[i, i]
            false_positive = conf_matrix[:, i].sum() - true_positive
            false_negative = conf_matrix[i, :].sum() - true_positive
            
            dice = (2 * true_positive) / (2 * true_positive + false_positive + false_negative)
            dice_scores[i] = dice
            
        return dice_scores

    plt.rcParams['font.weight'] = 'bold'
    plt.rcParams['axes.labelweight'] = 'bold'
    plt.rcParams['axes.titleweight'] = 'bold'
    
    def plot_evaluation_metrics(self, true_labels, predictions, features, conf_matrix, dice_scores, class_names):
        """Generate and save comprehensive evaluation metric plots with enhanced visualizations"""
        
        # 1. Confusion Matrix - Fixed overlapping y-axis labels
        plt.figure(figsize=(15, 12))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='viridis',
                    xticklabels=class_names, yticklabels=class_names)
        plt.title('Confusion Matrix', fontsize=14, fontweight='bold', pad=20)
        plt.xlabel('Predicted', fontsize=13, fontweight='bold')
        plt.ylabel('True', fontsize=12, fontweight='bold')
        
        # Fix overlapping labels by adjusting figure size and layout
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)  # Changed to horizontal labels
        ax = plt.gca()
        ax.set_xticklabels(ax.get_xticklabels(), fontsize=11, fontweight='bold')
        ax.set_yticklabels(ax.get_yticklabels(), fontsize=11, fontweight='bold')
        
        # Adjust layout to prevent label cutoff
        plt.tight_layout()
        plt.savefig('confusion_matrix.png', dpi=600, bbox_inches='tight')
        plt.close()
        
        # 2. Precision-Recall Curve
        plt.figure(figsize=(15, 10))
        y_true = np.eye(len(class_names))[true_labels]
        y_score = np.eye(len(class_names))[predictions]
        
        for i in range(len(class_names)):
            precision, recall, _ = precision_recall_curve(y_true[:, i], y_score[:, i])
            avg_precision = average_precision_score(y_true[:, i], y_score[:, i])
            plt.plot(recall, precision, label=f'{class_names[i]} (AP={avg_precision:.2f})', linewidth=2)
        
        plt.title('Precision-Recall Curve', fontsize=14, fontweight='bold', pad=20)
        plt.xlabel('Recall', fontsize=14, fontweight='bold')
        plt.ylabel('Precision', fontsize=14, fontweight='bold')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', prop={'size': 14, 'weight': 'bold'})
        plt.grid(True, linestyle='--', alpha=0.7)
        
        # Fix: Set tick label properties correctly
        ax = plt.gca()
        ax.set_xticklabels(ax.get_xticklabels(), fontsize=14, fontweight='bold')
        ax.set_yticklabels(ax.get_yticklabels(), fontsize=14, fontweight='bold')
        
        plt.tight_layout()
        plt.savefig('precision_recall.png', dpi=300, bbox_inches='tight')
        plt.close()
        
        # 3. t-SNE Visualization
        plt.figure(figsize=(15, 10))
        try:
            if len(features) > 50000:
                indices = np.random.choice(len(features), 50000, replace=False)
                features_subset = features[indices]
                labels_subset = true_labels[indices]
            else:
                features_subset = features
                labels_subset = true_labels
                
            tsne = TSNE(n_components=2, random_state=42, perplexity=30)
            features_embedded = tsne.fit_transform(features_subset)
            
            scatter = plt.scatter(features_embedded[:, 0], features_embedded[:, 1],
                                c=labels_subset, cmap='tab20', 
                                alpha=0.6, s=100, edgecolors='white', linewidth=0.5)
            
            cbar = plt.colorbar(scatter, ticks=range(len(class_names)))
            cbar.set_ticklabels(class_names)
            cbar.ax.tick_params(labelsize=14)
            for label in cbar.ax.get_yticklabels():
                label.set_fontweight('bold')
            
            plt.title('t-SNE Visualization of Features', fontsize=14, fontweight='bold', pad=20)
            plt.xlabel('t-SNE dimension 1', fontsize=14, fontweight='bold')
            plt.ylabel('t-SNE dimension 2', fontsize=14, fontweight='bold')
            plt.grid(True, linestyle='--', alpha=0.3)
            
            # Fix: Set tick label properties correctly
            ax = plt.gca()
            ax.set_xticklabels(ax.get_xticklabels(), fontsize=14, fontweight='bold')
            ax.set_yticklabels(ax.get_yticklabels(), fontsize=14, fontweight='bold')
            
        except Exception as e:
            plt.text(0.5, 0.5, f"t-SNE failed: {str(e)}", ha='center', va='center',
                    fontsize=14, fontweight='bold')
        
        plt.tight_layout()
        plt.savefig('tsne_visualization.png', dpi=300, bbox_inches='tight')
        plt.close()
        
        # 4. Dice Coefficients
        plt.figure(figsize=(15, 10))
        classes = list(range(len(dice_scores)))
        dice_values = list(dice_scores.values())
        
        plt.plot(classes, dice_values, marker='o', linewidth=2, markersize=8, 
                 color='blue', markerfacecolor='white', markeredgewidth=2)
        
        # Adjust vertical position of value labels to prevent overlap
        max_val = max(dice_values)
        for i, v in enumerate(dice_values):
            # Add more vertical space between point and label
            offset = (max_val * 0.04)  # Increased offset for better spacing
            plt.text(i, v + offset, f'{v:.2f}', ha='center', va='bottom',
                    fontsize=12, fontweight='bold')
        
        plt.title('Dice Coefficients by Class', fontsize=14, fontweight='bold', pad=20)
        plt.xlabel('Class', fontsize=14, fontweight='bold')
        plt.ylabel('Dice Coefficient', fontsize=14, fontweight='bold')
        
        plt.xticks(classes, class_names, rotation=45, ha='right')
        ax = plt.gca()
        ax.set_xticklabels(ax.get_xticklabels(), fontsize=12, fontweight='bold')
        ax.set_yticklabels(ax.get_yticklabels(), fontsize=12, fontweight='bold')
        
        plt.grid(True, linestyle='--', alpha=0.7)
        # Adjust y-axis limit to accommodate labels
        plt.ylim(0, max_val + (max_val * 0.15))  # Increased top margin
        
        plt.tight_layout()
        plt.savefig('dice_coefficients.png', dpi=300, bbox_inches='tight')
        plt.close()

    def plot_training_history(self):
        """Plot training history including both classification and nutrition metrics"""
        plt.figure(figsize=(20, 10))
        
        # Classification metrics
        plt.subplot(2, 2, 1)
        plt.plot(self.history['train_class_acc'], label='Train Accuracy')
        plt.plot(self.history['val_class_acc'], label='Validation Accuracy')
        if self.history['test_class_acc']:
            plt.axhline(y=self.history['test_class_acc'], color='r', linestyle='--', 
                       label='Test Accuracy')
        plt.title('Classification Accuracy', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch', fontsize=14, fontweight='bold')
        plt.ylabel('Accuracy (%)', fontsize=14, fontweight='bold')
        plt.legend(fontsize=14, prop={'weight': 'bold'})
        plt.xticks(fontsize=14, fontweight='bold')
        plt.yticks(fontsize=14, fontweight='bold')
        
        # Nutrition MSE
        plt.subplot(2, 2, 2)
        plt.plot(self.history['train_nutr_mse'], label='Train MSE')
        plt.plot(self.history['val_nutr_mse'], label='Validation MSE')
        if self.history['test_nutr_mse']:
            plt.axhline(y=self.history['test_nutr_mse'], color='r', linestyle='--',
                       label='Test MSE')
        plt.title('Nutrition Prediction MSE', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch', fontsize=14, fontweight='bold')
        plt.ylabel('MSE', fontsize=14, fontweight='bold')
        plt.legend(fontsize=14, prop={'weight': 'bold'})
        plt.xticks(fontsize=14, fontweight='bold')
        plt.yticks(fontsize=14, fontweight='bold')
        
        # Total loss
        plt.subplot(2, 2, 3)
        plt.plot(self.history['train_loss'], label='Train Loss')
        plt.plot(self.history['val_loss'], label='Validation Loss')
        plt.title('Total Loss', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch', fontsize=14, fontweight='bold')
        plt.ylabel('Loss', fontsize=14, fontweight='bold')
        plt.legend(fontsize=14, prop={'weight': 'bold'})
        plt.xticks(fontsize=14, fontweight='bold')
        plt.yticks(fontsize=14, fontweight='bold')
        
        plt.tight_layout()
        plt.savefig('training_history.png', dpi=300)
        plt.show()
        plt.close()
    def plot_nutrition_predictions(self, loader, dataset):
        """Generate comprehensive visualizations for nutrition predictions"""
        self.model.eval()
        true_nutrition = []
        pred_nutrition = []
        food_classes = []
        
        with torch.no_grad():
            for inputs, labels, nutrition in tqdm(loader, desc='Collecting predictions'):
                inputs = inputs.to(self.device)
                _, nutr_outputs = self.model(inputs)
                true_nutrition.extend(nutrition.numpy())
                pred_nutrition.extend(nutr_outputs.cpu().numpy())
                food_classes.extend([dataset.classes[label] for label in labels])
        
        true_nutrition = np.array(true_nutrition)
        pred_nutrition = np.array(pred_nutrition)
        
        # 1. Nutrient-wise Regression Plots
        fig, axes = plt.subplots(2, 3, figsize=(20, 12))
        axes = axes.ravel()
        
        for i, nutrient in enumerate(dataset.nutrition_columns):
            # Denormalize values
            true_vals = true_nutrition[:, i] * dataset.nutrition_stats[nutrient]['std'] + dataset.nutrition_stats[nutrient]['mean']
            pred_vals = pred_nutrition[:, i] * dataset.nutrition_stats[nutrient]['std'] + dataset.nutrition_stats[nutrient]['mean']
            
            # Calculate metrics
            r2 = r2_score(true_vals, pred_vals)
            mse = mean_squared_error(true_vals, pred_vals)
            
            # Create scatter plot with density heatmap
            sns.kdeplot(x=true_vals, y=pred_vals, ax=axes[i], cmap='viridis', fill=True)
            axes[i].scatter(true_vals, pred_vals, alpha=0.1, color='red', s=10)
            
            # Add diagonal line
            min_val = min(true_vals.min(), pred_vals.min())
            max_val = max(true_vals.max(), pred_vals.max())
            axes[i].plot([min_val, max_val], [min_val, max_val], 'r--', alpha=0.8)
            
            axes[i].set_title(f'{nutrient}\nR² = {r2:.3f}, MSE = {mse:.3f}', 
                             fontsize=14, fontweight='bold')
            axes[i].set_xlabel('True Values', fontsize=14, fontweight='bold')
            axes[i].set_ylabel('Predicted Values', fontsize=14, fontweight='bold')
            axes[i].tick_params(labelsize=14)
            axes[i].set_xticklabels(axes[i].get_xticklabels(), fontweight='bold')
            axes[i].set_yticklabels(axes[i].get_yticklabels(), fontweight='bold')
        
        plt.tight_layout()
        plt.savefig('nutrition_regression_plots.png', dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()
        
        # 2. Radar Chart of Prediction Accuracy by Nutrient
        mse_by_nutrient = []
        r2_by_nutrient = []
        
        for i in range(len(dataset.nutrition_columns)):
            true_vals = true_nutrition[:, i] * dataset.nutrition_stats[dataset.nutrition_columns[i]]['std'] + \
                       dataset.nutrition_stats[dataset.nutrition_columns[i]]['mean']
            pred_vals = pred_nutrition[:, i] * dataset.nutrition_stats[dataset.nutrition_columns[i]]['std'] + \
                       dataset.nutrition_stats[dataset.nutrition_columns[i]]['mean']
            
            mse_by_nutrient.append(mean_squared_error(true_vals, pred_vals))
            r2_by_nutrient.append(r2_score(true_vals, pred_vals))
        
        # Create radar chart
        angles = np.linspace(0, 2*np.pi, len(dataset.nutrition_columns), endpoint=False)
    
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7), subplot_kw=dict(projection='polar'))
        
        # MSE radar
        ax1.plot(angles, mse_by_nutrient, 'o-', linewidth=2)
        ax1.fill(angles, mse_by_nutrient, alpha=0.25)
        ax1.set_xticks(angles)
        ax1.set_xticklabels(dataset.nutrition_columns, fontsize=14)
        ax1.set_xticklabels(ax1.get_xticklabels(), fontweight='bold')
        
        # Format y-ticks to 2 decimal places
        ax1.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2f'))
        ax1.set_title('MSE by Nutrient', fontsize=14, fontweight='bold', pad=20)
        
        # R² radar with fixed decimal places
        ax2.plot(angles, r2_by_nutrient, 'o-', linewidth=2)
        ax2.fill(angles, r2_by_nutrient, alpha=0.25)
        ax2.set_xticks(angles)
        ax2.set_xticklabels(dataset.nutrition_columns, fontsize=14)
        ax2.set_xticklabels(ax2.get_xticklabels(), fontweight='bold')
        
        # Format y-ticks to 2 decimal places
        ax2.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2f'))
        ax2.set_title('R² Score by Nutrient', fontsize=14, fontweight='bold', pad=20)
        
        # Update yticks for both radar plots
        for ax in [ax1, ax2]:
            ax.tick_params(labelsize=14)
            ax.set_yticklabels([f'{float(label):.2f}' for label in ax.get_yticks()], 
                              fontweight='bold')
        
        plt.tight_layout()
        plt.savefig('nutrition_radar_charts.png', dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()
        
        # 3. Box plots of prediction errors by food class
        plt.figure(figsize=(15, 8))
        errors = []
        classes = []
        
        for i, nutrient in enumerate(dataset.nutrition_columns):
            true_vals = true_nutrition[:, i] * dataset.nutrition_stats[nutrient]['std'] + \
                       dataset.nutrition_stats[nutrient]['mean']
            pred_vals = pred_nutrition[:, i] * dataset.nutrition_stats[nutrient]['std'] + \
                       dataset.nutrition_stats[nutrient]['mean']
            rel_errors = np.abs((pred_vals - true_vals) / true_vals) * 100
            errors.extend(rel_errors)
            classes.extend([f"{nutrient}-{cls}" for cls in food_classes])
        
        sns.boxplot(x='Nutrient-Food Class', y='Relative Error (%)', 
                    data=pd.DataFrame({'Nutrient-Food Class': classes, 'Relative Error (%)': errors}))
        plt.xticks(rotation=45, ha='right', fontsize=14)
        plt.gca().set_xticklabels(plt.gca().get_xticklabels(), fontweight='bold')
        plt.yticks(fontsize=14)
        plt.gca().set_yticklabels(plt.gca().get_yticklabels(), fontweight='bold')
        plt.xlabel('Nutrient-Food Class', fontsize=14, fontweight='bold')
        plt.ylabel('Relative Error (%)', fontsize=14, fontweight='bold')
        plt.title('Distribution of Prediction Errors by Food Class and Nutrient', 
                 fontsize=14, fontweight='bold', pad=20)
        plt.tight_layout()
        plt.savefig('nutrition_error_distribution.png', dpi=300, bbox_inches='tight')
        plt.close()
        
        # 4. Interactive 3D scatter plot of nutrient predictions
        fig = plt.figure(figsize=(12, 8))
        ax = fig.add_subplot(111, projection='3d')
        
        # Select three nutrients for 3D visualization
        nutrients = dataset.nutrition_columns[:3]
        
        # Denormalize values for selected nutrients
        true_vals = np.array([true_nutrition[:, i] * dataset.nutrition_stats[nutrients[i]]['std'] + 
                             dataset.nutrition_stats[nutrients[i]]['mean'] for i in range(3)])
        pred_vals = np.array([pred_nutrition[:, i] * dataset.nutrition_stats[nutrients[i]]['std'] + 
                             dataset.nutrition_stats[nutrients[i]]['mean'] for i in range(3)])
        
        scatter = ax.scatter(true_vals[0], true_vals[1], true_vals[2], 
                            c=pred_vals[0], cmap='viridis', alpha=0.6)
        
        ax.set_xlabel(nutrients[0], fontsize=14, fontweight='bold')
        ax.set_ylabel(nutrients[1], fontsize=14, fontweight='bold')
        ax.set_zlabel(nutrients[2], fontsize=14, fontweight='bold')
        
        # Update colorbar
        cbar = plt.colorbar(scatter)
        cbar.set_label(f'Predicted {nutrients[0]}', fontsize=14, fontweight='bold')
        cbar.ax.tick_params(labelsize=14)
        cbar.ax.set_yticklabels(cbar.ax.get_yticklabels(), fontweight='bold')
        
        # Update axis ticks
        ax.tick_params(labelsize=14)
        for axis in [ax.xaxis, ax.yaxis, ax.zaxis]:
            axis.set_ticklabels(axis.get_ticklabels(), fontweight='bold')
        
        plt.title('3D Nutrient Prediction Visualization', 
                 fontsize=14, fontweight='bold', pad=20)
        
        plt.tight_layout()
        plt.savefig('nutrition_3d_visualization.png', dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()
def predict_and_get_nutrition(image_path, classifier, dataset):
    """Predict food class and nutrition values directly from image"""
    # Prepare image
    image = Image.open(image_path).convert('RGB')
    input_tensor = classifier.val_transform(image).unsqueeze(0).to(classifier.device)
    
    # Get predictions
    classifier.model.eval()
    with torch.no_grad():
        class_output, nutrition_output = classifier.model(input_tensor)
        class_probs = torch.softmax(class_output, dim=1)
        confidence, predicted_idx = class_probs.max(1)
    
    # Convert class index to name
    idx_to_class = {v: k for k, v in dataset.class_to_idx.items()}
    predicted_class = idx_to_class[predicted_idx.item()]
    
    # Denormalize nutrition values
    nutrition_pred = nutrition_output[0].cpu().numpy()
    denormalized_nutrition = dataset.denormalize_nutrition(nutrition_pred)
    
    return {
        'food_class': predicted_class,
        'confidence': confidence.item() * 100,
        'nutrition_values': denormalized_nutrition
    }

def visualize_random_predictions(classifier, test_loader, dataset, num_samples=5):
    """
    Visualize predictions for random images from the test set
    
    Args:
        classifier: Trained FoodClassifier instance
        test_loader: DataLoader containing test data
        dataset: FoodDataset instance
        num_samples: Number of random samples to visualize
    """
    import random
    import torchvision.transforms as transforms
    import matplotlib.pyplot as plt
    
    # Set model to evaluation mode
    classifier.model.eval()
    
    # Get random batch
    data_iter = iter(test_loader)
    batch = next(data_iter)
    images, labels, nutrition = batch
    
    # Select random indices
    batch_size = len(images)
    indices = random.sample(range(batch_size), min(num_samples, batch_size))
    
    # Create figure
    fig = plt.figure(figsize=(20, 4*num_samples))
    
    # Inverse transform for displaying images
    inverse_transform = transforms.Compose([
        transforms.Normalize(
            mean=[-0.485/0.229, -0.456/0.224, -0.406/0.225],
            std=[1/0.229, 1/0.224, 1/0.225]
        )
    ])
    
    # Create mapping from index to class name
    idx_to_class = {v: k for k, v in dataset.class_to_idx.items()}
    
    with torch.no_grad():
        for i, idx in enumerate(indices):
            # Get image and true label
            image = images[idx].unsqueeze(0).to(classifier.device)
            true_label = labels[idx].item()
            true_nutrition = nutrition[idx]
            
            # Get predictions
            class_output, nutrition_output = classifier.model(image)
            class_probs = torch.softmax(class_output, dim=1)
            confidence, predicted_idx = class_probs.max(1)
            
            # Denormalize nutrition values
            true_nutrition_denorm = dataset.denormalize_nutrition(true_nutrition.numpy())
            pred_nutrition_denorm = dataset.denormalize_nutrition(nutrition_output[0].cpu().numpy())
            
            # Display image
            ax1 = plt.subplot(num_samples, 2, 2*i + 1)
            img_display = inverse_transform(images[idx]).permute(1, 2, 0).cpu().numpy()
            img_display = np.clip(img_display, 0, 1)
            plt.imshow(img_display)
            plt.axis('off')
            
            # Add prediction text
            title = f'Predicted: {idx_to_class[predicted_idx.item()]}\n'
            title += f'True: {idx_to_class[true_label]}\n'
            title += f'Confidence: {confidence.item()*100:.2f}%'
            plt.title(title, pad=10)
            
            # Display nutrition comparison
            ax2 = plt.subplot(num_samples, 2, 2*i + 2)
            nutrients = list(true_nutrition_denorm.keys())
            x = np.arange(len(nutrients))
            width = 0.35
            
            true_values = [true_nutrition_denorm[n] for n in nutrients]
            pred_values = [pred_nutrition_denorm[n] for n in nutrients]
            
            ax2.bar(x - width/2, true_values, width, label='True', color='green', alpha=0.6)
            ax2.bar(x + width/2, pred_values, width, label='Predicted', color='blue', alpha=0.6)
            
            ax2.set_xticks(x)
            ax2.set_xticklabels(nutrients, rotation=45, ha='right')
            ax2.legend()
            ax2.set_title('Nutrition Values Comparison')
            
            # Add relative error text
            for j, (true_val, pred_val) in enumerate(zip(true_values, pred_values)):
                rel_error = abs(true_val - pred_val) / true_val * 100
                ax2.text(j, max(true_val, pred_val), f'{rel_error:.1f}%', 
                        ha='center', va='bottom', color='red')
    plt.tight_layout()
    plt.savefig('visual prediction', dpi=300, bbox_inches='tight')
    plt.show()

# Example usage:
# visualize_random_predictions(classifier, test_loader, dataset)

def main():
    """Main function to run the training or prediction pipeline"""
    # Configuration
    data_root = '/kaggle/input/food-antioxidant-cv-dataset/FOOD DATASET'  # Update with your dataset path
    csv_file = '/kaggle/input/food-antioxidant-csv-file/food_antioxidant.csv'  # Update with your CSV file path
    batch_size = 16
    num_epochs = 15
    
    # Initialize dataset
    full_dataset = FoodDataset(data_root, csv_file, transform=None, samples_per_class=1000)
    num_classes = len(full_dataset.class_to_idx)
    
    # Split indices for train/val/test
    indices = list(range(len(full_dataset)))
    train_idx, temp_idx = train_test_split(indices, test_size=0.3, random_state=42)
    val_idx, test_idx = train_test_split(temp_idx, test_size=0.5, random_state=42)
    
    # Choose mode
    mode = input("Enter mode (train/predict): ").lower()
    
    if mode == 'train':
        # Create datasets with appropriate transforms
        train_dataset = FoodDataset(data_root, csv_file, transform=None)
        val_dataset = FoodDataset(data_root, csv_file, transform=None)
        test_dataset = FoodDataset(data_root, csv_file, transform=None)
        
        # Initialize classifier
        classifier = FoodClassifier(num_classes=num_classes)
        
        # Set transforms
        train_dataset.transform = classifier.train_transform
        val_dataset.transform = classifier.val_transform
        test_dataset.transform = classifier.val_transform
        
        # Create data loaders
        train_loader = DataLoader(
            train_dataset, batch_size=batch_size,
            sampler=torch.utils.data.SubsetRandomSampler(train_idx),
            num_workers=4, pin_memory=True
        )
        val_loader = DataLoader(
            val_dataset, batch_size=batch_size,
            sampler=torch.utils.data.SubsetRandomSampler(val_idx),
            num_workers=4, pin_memory=True
        )
        test_loader = DataLoader(
            test_dataset, batch_size=batch_size,
            sampler=torch.utils.data.SubsetRandomSampler(test_idx),
            num_workers=4, pin_memory=True
        )
        
        # Train the model
        print("Starting training...")
        classifier.train_model(train_loader, val_loader, test_loader, num_epochs)
        print("Training completed. Model saved as 'best_model.pth'")
        
    elif mode == 'predict':
        try:
            # Initialize dataset and classifier
            test_dataset = FoodDataset(data_root, csv_file, transform=None)
            num_classes = len(test_dataset.class_to_idx)
            classifier = FoodClassifier(num_classes=num_classes)
            
            # Set transform
            test_dataset.transform = classifier.val_transform
            
            # Create test loader
            test_loader = DataLoader(
                test_dataset, batch_size=batch_size,
                sampler=torch.utils.data.SubsetRandomSampler(test_idx),
                num_workers=4, pin_memory=True
            )
            
            # Load saved model
            model_path = 'best_model.pth'
            classifier.load_saved_model(model_path)
            
            # Visualization option
            vis_option = input("Choose visualization mode (1: Single image, 2: Random test samples): ")
            
            if vis_option == '1':
                # Get image path from user
                image_path = input("Enter the path to your food image: ")
                
                # Make prediction
                results = predict_and_get_nutrition(image_path, classifier, test_dataset)
                
                # Display results
                print("\nPrediction Results:")
                print(f"Predicted food: {results['food_class']}")
                print(f"Confidence: {results['confidence']:.2f}%")
                print("\nNutrition Information:")
                for key, value in results['nutrition_values'].items():
                    print(f"{key}: {value:.2f}")
            
            elif vis_option == '2':
                # Visualize random test samples
                print("\nVisualizing random test samples...")
                visualize_random_predictions(classifier, test_loader, test_dataset)
            
            else:
                print("Invalid visualization option.")
                
        except FileNotFoundError as e:
            print(f"Error: Could not find file - {str(e)}")
        except Exception as e:
            print(f"An error occurred: {str(e)}")
            raise  # Add this to see the full error traceback
            
    else:
        print("Invalid mode. Please choose 'train' or 'predict'")

if __name__ == '__main__':
    main()
```
